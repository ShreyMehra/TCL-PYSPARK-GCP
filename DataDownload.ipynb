```python
from google.cloud import storage
from google.oauth2 import service_account
import requests
import time

bucket_name = "training-project-404110.appspot.com"  
storage_client = storage.Client()
bucket = storage_client.bucket(bucket_name)

def putfile(url_a, url_b):
    try:
        url = url_a + url_b
        response = requests.get(url)
        
        blob_name = f"Data/{url_b}"
        blob = bucket.blob(blob_name)

        blob.upload_from_string(response.content)

    except Exception as e:
        print("FUNC Error", e)

def upload():
    for year in range(2009, 2024):
        for month in range(1, 13):
            try:
                url_a = 'https://d37ci6vzurychx.cloudfront.net/trip-data/'
                url_b = f'yellow_tripdata_{year}-{month:02}.parquet'
                putfile(url_a, url_b)

                print(f"File uploaded for {url_b} {time.ctime()}")
            except Exception as e:
                print("FOR Error", e)

upload()
```
